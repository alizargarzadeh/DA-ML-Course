```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure for the plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_axes([0, 0, 1, 1])


# Create a Basemap object to define the map projection and boundaries
m = Basemap(llcrnrlon=42., llcrnrlat=25., urcrnrlon=64, urcrnrlat=42.,
            rsphere=(6378137.00, 6356752.3142),
            resolution='h')

# Define information about cities including name, coordinates, and population
tehran = {"name": "Tehran", "x": 51, "y": 35, "population": 7153309}
mashhad = {"name": "Mashhad", "x": 59, "y": 36, "population": 2307177}
isfahan = {"name": "Isfahan", "x": 51, "y": 32, "population": 1547164}
karaj = {"name": "Karaj", "x": 50, "y": 35, "population": 1448075}
shiraz = {"name": "Shiraz", "x": 52, "y": 29, "population": 1249942}
tabriz = {"name": "Tabriz", "x": 46, "y": 38, "population": 1424641}
qom = {"name": "Qom", "x": 50, "y": 34, "population": 900000}
ahvaz = {"name": "Ahvaz", "x": 48, "y": 31, "population": 841145}
kermanshah = {"name": "Kermanshah", "x": 47, "y": 34, "population": 621100}
urmia = {"name": "Urmia", "x": 45, "y": 37, "population": 577307}

city = [tehran, mashhad, isfahan, karaj, shiraz,
        tabriz, qom, ahvaz, kermanshah, urmia]

# Draw map features
m.drawcoastlines()
m.drawcountries()
m.drawmapboundary(fill_color='aqua')
m.fillcontinents(color='white', lake_color='aqua')
m.drawparallels(np.arange(25, 41, 5), labels=[1, 1, 0, 0])
m.drawmeridians(np.arange(40, 65, 5), labels=[0, 0, 0, 1])


# Plot city locations with their populations as point sizes
for i in city:
    plt.scatter(i["x"], i["y"], label=i["name"],
                s=i["population"]/5000, alpha=0.5)


# Add a legend to the plot
plt.legend(loc=1)

# Set the title for the plot
ax.set_title("Population of Cities in Iran 2023")

# Display the plot
plt.show()
```
